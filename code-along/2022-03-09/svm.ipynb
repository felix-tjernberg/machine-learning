```python
from utils import plot_svm_margins
from sklearn.svm import SVC
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


blobs_raw = make_blobs([300, 300], 2, centers=[(-2, -2), (2, 2)], random_state=42)
blobs_raw
```

```python
blobs = blobs_raw[0]
blobs
```

```python
blobs[:, 0]
```

```python
plt.scatter(blobs[:, 0], blobs[:, 1])
```

## Clustering
- K-means algorithm
- unsupervised learning algorithm

```python
from sklearn.cluster import KMeans

kmean = KMeans(2)
kmean.fit(blobs)
kmean.predict(blobs)
```

```python
blobs_predicted = np.c_[blobs, kmean.predict(blobs)]
blobs_predicted
```

```python
df_blobs = pd.DataFrame(blobs_predicted, columns = ["X1", "X2", "label"])
sns.scatterplot(data = df_blobs, x = "X1", y = "X2", hue = "label")
```

```python
X, y = df_blobs.drop("label", axis = 1).to_numpy(), df_blobs["label"].to_numpy()
X
```

## SVM

### Linear kernel

```python
fig, axes = plt.subplots(1,3, dpi=100, figsize=(12,4))

for index, C in enumerate([10,.1,.01], 1):
    plt.subplot(1,3,index)
    classifier = SVC(kernel='linear', C=C)
    plot_svm_margins(classifier, X, y)
```

### Radial basis kernel

```python
fig, axes = plt.subplots(1,3, dpi=100, figsize=(12,4))

for index, C in enumerate([10,.1,.01], 1):
    plt.subplot(1,3,index)
    classifier = SVC(kernel='rbf', C=C)
    plot_svm_margins(classifier, X, y)
```

### Polynomial kernel

```python
fig, axes = plt.subplots(1,3, dpi=100, figsize=(12,4))

for index, C in enumerate([10,.1,.01], 1):
    plt.subplot(1,3,index)
    classifier = SVC(kernel='poly', C=C, degree=3)
    plot_svm_margins(classifier, X, y)
```

### Sigmoid kernel

```python
fig, axes = plt.subplots(1,3, dpi=100, figsize=(12,4))

for index, C in enumerate([10,.1,.01], 1):
    plt.subplot(1,3,index)
    classifier = SVC(kernel='sigmoid', C=C)
    plot_svm_margins(classifier, X, y)
```
